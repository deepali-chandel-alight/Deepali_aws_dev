```python
def find_files_ext_dist(bucket_name, source_folder_key):
    s3 = boto3.resource('s3')
    bucket = s3.Bucket(bucket_name)
    
    ext_count_dict = {}
    
    for object_summary in bucket.objects.filter(Prefix=source_folder_key):
        _, file_ext = os.path.splitext(object_summary.key)
        try:
            ext_count_dict[file_ext] +=1
        except:
            ext_count_dict[file_ext] = 1
            
    return pd.DataFrame(ext_count_dict.items(), columns=['file_ext', 'file_count']
                       ).set_index('file_ext').sort_values(by='file_count', ascending=False)

            
df = find_files_ext_dist(bucket_name, source_folder_key)
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a Pandas Series
series = df.file_count

# Calculate the total sum of the series
total_sum = series.sum()

# Find categories with less than 5% of the total and merge them into "Others"
threshold = 0.01  # 5%
small_categories = series[series / total_sum < threshold]
series['Others'] = small_categories.sum()
series.drop(small_categories.index, inplace=True)

# Create a pie chart
plt.figure(figsize=(5, 5))
plt.pie(series, labels=series.index, autopct='%1.1f%%', startangle=140, wedgeprops=dict(width=0.3))
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.

# Add a title
plt.title("Distribution of Categories")

# Display the chart
plt.show()
```

```python
# Initialize S3 client
s3 = boto3.client('s3')

def write_pdf(list_paragraphs, bucket_name, output_pdf_key):
    try:
        # Create a list to hold the paragraphs and spacers
        story = []

        # Create a style for the paragraphs
        styles = getSampleStyleSheet()
        normal_style = styles['Normal']
        for text in list_paragraphs:
            if text:  # Only add non-empty paragraphs
                text = text.strip()  # Remove leading and trailing spaces
                paragraph = Paragraph(text, normal_style)
                story.append(paragraph)
                story.append(Spacer(1, 12))  # Add some space between paragraphs
        
        # Create a PDF in memory using ReportLab
        pdf_buffer = io.BytesIO()
        doc = SimpleDocTemplate(pdf_buffer, pagesize=letter)
        
        # Build the PDF document
        doc.build(story)
        
        # Upload the generated PDF to S3
        s3.put_object(Bucket=bucket_name, Key=output_pdf_key, Body=pdf_buffer.getvalue())
    except Exception as e:
        print("Error:", e)

def docx_to_pdf(docx_content, bucket_name, pdf_path):
    list_paragraphs=[]
    try:
        # Create a Document object from the .docx content
        doc = Document(io.BytesIO(docx_content))
        #print(len(doc.paragraphs))
        for para in doc.paragraphs:
            list_paragraphs.append(para.text)
    except Exception as e:
        print("Error:", e)
        raise
        
    # Write extracted text to pdf
    if list_paragraphs:
        write_pdf(list_paragraphs, bucket_name, pdf_path)
#----------------------------------------

def pptx_to_pdf(pptx_content, bucket_name, pdf_path):
    list_paragraphs=[]
    try:
        # Load PPT/PPTX data into a PowerPoint presentation object
        ppt = Presentation(BytesIO(pptx_content))
        # Iterate through slides and add text to PDF
        for slide in ppt.slides:
            text = ''
            for shape in slide.shapes:
                if shape.has_text_frame:
                    text += shape.text + '\n'
            list_paragraphs.append(text)
    except Exception as e:
        print("Error:", e)
        raise
        
    # Write extracted text to pdf
    if list_paragraphs:
        write_pdf(list_paragraphs, bucket_name, pdf_path)


def msg_to_pdf(msg_content, bucket_name, pdf_path):
    list_paragraphs=[]
    try:  
        # Create a Message object from the downloaded content
        msg = Message(msg_content)
            
        # Get the subject, sender, and body from the .msg file
        list_paragraphs.append(f"Subject: {msg.subject}")
        list_paragraphs.append(f"Sender: {msg.sender}")
        list_paragraphs.append("Body:") 
        if isinstance(msg.body, str):
            list_paragraphs.extend(msg.body.split('\n'))
    except Exception as e:
        print(f"Error reading .msg file from S3: {e}")
        raise
    
    # Write extracted text to pdf
    if list_paragraphs:
        write_pdf(list_paragraphs, bucket_name, pdf_path)

def txt_to_pdf(txt_content, bucket_name, pdf_path):
    list_paragraphs=[]
    try:  
        content = txt_content.decode('utf-8')
        list_paragraphs.append(content)
    except Exception as e:
        print(f"Error reading .msg file from S3: {e}")
        raise
        
    # Write extracted text to pdf
    if list_paragraphs:
        write_pdf(list_paragraphs, bucket_name, pdf_path)
    
# convert .xlsx, xls, to .csv
def excel_to_csv(excel_data, bucket_name, pdf_path):
    try:
        output_file_key = pdf_path[:pdf_path.rfind(".pdf")] + "." + '.csv'

        # Load Excel data into a pandas DataFrame
        excel_df = pd.read_excel(BytesIO(excel_data))

        # Convert DataFrame to CSV format
        csv_data = excel_df.to_csv(index=False)

        # Upload the CSV data to S3
        s3.put_object(Bucket=bucket_name, Key=output_file_key, Body=csv_data.encode('utf-8'))
    except Exception as e:
        print(f"Error reading excel file from S3: {e}")
        raise
        
def doc_to_pdf(doc_content, bucket_name, pdf_path):
    list_paragraphs=[]
    try:
        # Create a seekable file-like object from the bytes content
        doc_stream = io.BytesIO(doc_content)   

        # Load the Word document using python-docx
        doc = docx.Document(doc_stream)

        # Extract text from paragraphs and save it as a text file
        text_content = "\n".join([paragraph.text for paragraph in doc.paragraphs])
        print(text_content)
    except Exception as e:
        print("Error:", e)
        raise
        
    # Write extracted text to pdf
    if list_paragraphs:
        write_pdf(list_paragraphs, bucket_name, pdf_path)

def ppt_to_pdf(ppt_content, bucket_name, pdf_path):
    list_paragraphs=[]
    try:
        # Load the PowerPoint presentation using python-pptx
        presentation = Presentation(io.BytesIO(ppt_content))

        # Extract text from slides and save it as a text file
        text_content = ""
        for slide in presentation.slides:
            for shape in slide.shapes:
                if hasattr(shape, "text"):
                    text_content += shape.text + "\n"
        print(ppt_content)
    except Exception as e:
        print("Error:", e)
        raise
        
    # Write extracted text to pdf
    if list_paragraphs:
        write_pdf(list_paragraphs, bucket_name, pdf_path)
```

```python
# Convert each file to PDF
def convert_to_pdf_util(bucket_name, input_file_key, pdf_path):
    base_name = os.path.basename(input_file_key)
    file_name, file_ext = os.path.splitext(base_name)
    pdf_path, _ = os.path.splitext(pdf_path)
    pdf_path = f"{pdf_path}.pdf"
    
    # read file content in binary
    response = s3.get_object(Bucket=bucket_name, Key=input_file_key)
    file_content = response['Body'].read()
    
    # Convert binary content to assci content(normal text) based on specific input file ext 
    if file_ext in ['.xlsx', '.xls']:
        #convert and save as csv
        excel_to_csv(file_content, bucket_name, pdf_path)
    elif file_ext in ['.doc']:
        doc_to_pdf(file_content, bucket_name, pdf_path)
    elif file_ext in ['.ppt']:
        ppt_to_pdf(file_content, bucket_name, pdf_path)
    elif file_ext in ['.pptx']:
        pptx_to_pdf(file_content, bucket_name, pdf_path)
    elif file_ext in ['.docx']: 
        docx_to_pdf(file_content, bucket_name, pdf_path)    
    elif file_ext in ['.msg']: 
        msg_to_pdf(file_content, bucket_name, pdf_path)
    elif file_ext == '.txt':
        txt_to_pdf(file_content, bucket_name, pdf_path)
    

def convert_to_pdf(bucket_name, source_folder_key, dest_folder_key):
    s3_resource = boto3.resource('s3')
    bucket = s3_resource.Bucket(bucket_name)
    selected_ext = (".doc", ".docx", ".xls", ".xlsx", ".msg", ".txt", ".ppt", ".pptx")
    # selected_ext = (".doc", ".ppt")
    
    print(f"From <{source_folder_key}>, Converting <{selected_ext}> files to pdf & saving to <{dest_folder_key}>")
    print('-'*40)
    
    # num_files_copied_directly = 0
    num_files_processed = 0
    num_files_failed = 0
    failed_files = []
    
    num_files_copied_directly = {ext: 0 for ext in ('.pdf', '.csv')}
    ext_processed_counts = {ext: 0 for ext in selected_ext}
    ext_failed_files = {ext: [] for ext in selected_ext}
    
    for object_summary in bucket.objects.filter(Prefix=source_folder_key):
        key = object_summary.key
        dest_key = dest_folder_key + key[len(source_folder_key):]
        
        # Copy file directly if its ('.pdf', '.csv')
        if key.endswith(('.pdf', '.csv')):
            source_obj = {'Bucket': bucket_name, 'Key': key}
            try:
                s3_resource.meta.client.copy(source_obj, bucket_name, dest_key)
                #print(f"Copied {key} to {dest_key}")
                ext = key[key.rfind('.'):].lower()
                num_files_copied_directly[ext] += 1
                #print()
            except Exception as e:
                print(f"Error copying {key}: {e}")
                failed_files.append(key)
                num_files_failed += 1
                print()

        if key.endswith(selected_ext):
            try:
                # Convert and keep folder structure intact
                convert_to_pdf_util(bucket_name, key, dest_key)
                #print(f"Processed: <{key}>")
                ext = key[key.rfind('.'):].lower()  # Extract extension and convert to lowercase
                ext_processed_counts[ext] += 1
                num_files_processed += 1
            except Exception as e:
                print(f"Error Processing: <{key}> - {e}")
                failed_files.append(key)
                ext = key[key.rfind('.'):].lower()  # Extract extension and convert to lowercase
                ext_failed_files[ext].append(key)
                num_files_failed += 1
                print()
    
    print('='*80)
    print("Conversion Summary:")
    print(f"Files copied directly: {num_files_copied_directly}")
    print(f"Files successfully processed: {num_files_processed}")
    print(f"Files failed to process: {num_files_failed}")
    print("\nFiles processed per extension:")
    for ext, count in ext_processed_counts.items():
        print(f"{ext}: {count} processed, {len(ext_failed_files[ext])} failed")
    print('='*80)
    print("\nFailed files per extension:")
    for ext, failed_list in ext_failed_files.items():
        #print(f"{ext}: {failed_list}\n")
        print(f"{ext}:")
        for file in failed_list:
            print(file)
        print('-'*80)
    
    # return failed_files
    return num_files_copied_directly, num_files_processed, num_files_failed, ext_processed_counts, ext_failed_files
```

```python
# Example usage
num_files_copied_directly, num_files_processed, num_files_failed, ext_processed_counts, ext_failed_files = convert_to_pdf(bucket_name, source_folder_key, dest_folder_key)
```
